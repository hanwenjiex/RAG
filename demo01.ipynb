```python
from clean_document import read_word_document, parse_document_structure, extract_document_structure, flatten_structure, clean_document
import json
docx_path = "综合能源数字化管控平台技术方案20230821.docx"
# 1. 读取 Word 文档
raw_text = read_word_document(docx_path)
print("读取文档成功。")

sections = parse_document_structure(docx_path)

structure = extract_document_structure(docx_path)
#print(json.dumps(structure, ensure_ascii=False, indent=2))

# 解析structure 
knowledge_struct = []
knowledge_content = []
head = structure[0].get('title') + '-' +  structure[0].get('content') + "包括以下的框架：" 
for i in range(1, len(structure)):
	if i == len(structure) - 1:
		head = head + structure[i].get('title')
	else:
		head = head + structure[i].get('title') + '、'
head = head + '。'
structure.pop(0)
knowledge_struct.append(head)
# 扁平化结构
flattened_data, content_flattened_data = flatten_structure(structure)
# 输出扁平化后的列表数据
for item in flattened_data:
	tmp_knowledge = item[0] + "的内容包括：" 
	for item_i in item[1:]:
		if item_i == item[-1]:
			tmp_knowledge = tmp_knowledge + item_i
		else:
			tmp_knowledge = tmp_knowledge + item_i + "、"
	cleaned_text = clean_document(tmp_knowledge)	
	knowledge_struct.append(cleaned_text)

for item in content_flattened_data:
	tmp_knowledge = item[0] + "的内容包括：" 
	for item_i in item[1:]:
		if item_i == item[-1]:
			tmp_knowledge = tmp_knowledge + item_i
		else:
			tmp_knowledge = tmp_knowledge + item_i + "、"
	cleaned_text = clean_document(tmp_knowledge)	
	knowledge_content.append(cleaned_text)


print(knowledge_struct)

print("--------------------")

print(knowledge_content)
```

```python
# 2. jeix遍历每个子节点的，生成内容

file_contents = json.loads(first_content)
def find_empty_nodes(data, path=""):
    """
    遍历 JSON 数据结构，找出值为空的节点，并记录下路径。
    
    :param data: JSON 数据结构（可以是字典、列表、字符串等）
    :param path: 当前节点的路径（用于递归时跟踪位置）
    :return: 返回所有空节点的路径列表
    """
    empty_nodes = []

    if isinstance(data, dict):  # 如果是字典，递归遍历每个键值对
        for key, value in data.items():
            new_path = f"{path}@{key}" if path else key  # 构建当前节点的路径
            empty_nodes.extend(find_empty_nodes(value, new_path))  # 递归调用

    elif isinstance(data, list):  # 如果是列表，递归遍历每个元素
        for idx, value in enumerate(data):
            new_path = f"{path}[{idx}]"
            empty_nodes.extend(find_empty_nodes(value, new_path))  # 递归调用

    else:  # 对于字典和列表之外的数据类型，检查是否为空
        if data == "" or data is None or data == [] or data == {}:
            empty_nodes.append(path)  # 记录空节点的路径

    return empty_nodes

# 查找空节点
empty_nodes = find_empty_nodes(file_contents)

def assign_value_to_empty_node(data, path, value):
    """
    根据路径回溯并给空节点赋值。
    
    :param data: JSON 数据结构
    :param path: 需要赋值的节点路径
    :param value: 赋予空节点的值
    """
    keys = path.lstrip('@').split('@')  # 将路径分解为键列表
    node = data
    # 逐级回溯路径
    for key in keys[:-1]:  # 遍历路径中的每一部分，直到最后一个节点
        if isinstance(node, dict):
            node = node.get(key, {})  # 如果是字典，获取键值对
        elif isinstance(node, list):
            idx = int(key.strip('[]'))  # 获取列表索引
            if idx < len(node):
                node = node[idx]  # 获取列表中的元素
            else:
                node = {}  # 如果索引超出范围，返回空字典
                
    
    
    
    
    # 最后一个节点赋值
    last_key = keys[-1]
    if isinstance(node, dict):
        node[last_key] = value  # 如果是字典，直接赋值
    elif isinstance(node, list):
        idx = int(last_key.strip('[]'))  # 获取索引
        if 0 <= idx < len(node):
            node[idx] = value  # 如果是列表，给对应索引位置赋值

# 输出结果
print("空节点路径：")
for node in empty_nodes:
    print(node)
 

# 给空节点赋值
for path in empty_nodes:
    keys = path.lstrip('@').split('@')   
    content_name = keys[-1]
    print("正在生成" + content_name)
    content_query = query + content_name
    retrieved_chunks_content = retrieve_chunks(content_query, embed_model_content, index, chunks_content, top_k=10)
    prompt_content = "请参考以下信息：\n"
    for idx, chunk in enumerate(retrieved_chunks_content, 1):
	    prompt_content += f"[参考 {idx}]: {chunk}\n"
    prompt_content += f"用户问题：“{query}”，现在要求你基于参考信息仅生成该问题关于“{content_name}”的内容, 字数在100-500字之间\n"
    
    completion = client.chat.completions.create(
        model="qwen-plus",  # 
        messages=[{'role': 'user', 'content': prompt_content}],)
    
    tmp_content = completion.choices[0].message.content
    assign_value_to_empty_node(file_contents, path, tmp_content)  # 这里赋值为 "默认值"
    
print("\n修改后的 JSON 数据：")
print(json.dumps(file_contents, ensure_ascii=False, indent=2))
print(file_contents)
```
